1. Given that 1 minute of latitude is approximately equal to 1 nautical mile (1852 meters),
how many significant digits after the decimal must be included for a latitude represented
in degrees to describe a fix that is accurate to 1 cm? How many significant digits are required after the decimal in the arc-seconds field if the latitude is represented in degrees,
arc-minutes, and arc-seconds to describe a fix that is accurate to 1 cm? Note: 1 degree =
60 arc-minutes, 1 arc-minute = 60 arc-seconds (you may find arc-minutes referred to as
‘minutes’ and arc-seconds referred to as ‘seconds’).


```python
print(1 / 1852 / 100)  # degrees for 1 cm
print((1 / 1852 / 100) * 3600)  # arcseconds for 1 cm
```

$$
\frac{1}{1852} \frac{\text{degrees}}{\text{meter}} \cdot \frac{1}{100} \frac{\text{m}}{\text{cm}} = 5.4 \times 10^{-6} \frac{\text{degrees}}{\text{cm}}
$$

$\Rightarrow$ 7 significant digits after the decimal in degrees.

$$
5.4 \times 10^{-6} \frac{\text{degrees}}{\text{cm}} \cdot 3600 \frac{\text{arcseconds}}{\text{degree}} = 1.94\times 10^{-2} \frac{\text{arcseconds}}{\text{cm}}
$$

$\Rightarrow$ 3 significant digits after the decimal in arcseconds.


2. Suppose you start from location 45° N, 120° W (lat, long) and fly at an altitude of 10
km with ground speed of 885 km/h for eight hours at a constant heading of 45° from
true north. Where would you end up? Assume a spherical earth with radius of 6371 km.
Two helpful notes:
- Ground speed in aviation means the speed of an aircraft relative
to the surface of the earth (to be distinguished from air speed, which means the speed
of an aircraft relative to its surrounding air mass).
- It’s safe to say that you’ll end up
pretty far north.

Note that only the ground speed is given, so the aircraft altitude is irrelevant to the problem.  If we decompose the velocity $v$ into its north ($v_N$) and east ($v_E$) components, we find:

$$
v_N = v_E = \frac{\sqrt{2}}{2} v
$$

Since the aircraft travels for a total of 8 hours, its total northward displacement is $4\sqrt{2} \cdot 885 \approx $ 5006 km.

```python
import numpy as np
print(4 * np.sqrt(2) * 885)
```

Note that the starting location, which is at 45° N, is $\frac{1}{8}$-th the Earth's circumference $2\pi r_E$ from the North pole.  Thus, the aircraft only needs to travel a ground distance of $\frac{\pi r_E}{4} \approx$ 5004 km.

Since the northward displacement (approx. 5006 km) is greater than the distance from the north pole (approx. 5004 km) the aircraft will have reached the north pole -- i.e. 90° N.

Aside: longitude is undefined at +/- 90° latitude.  This problem shows that the idea of having a constant east/north heading becomes troublesome near the poles.  It is good motivation for having other coordinate systems with which to work.
